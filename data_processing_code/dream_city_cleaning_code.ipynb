# **SCHOOLS**

```python
#read in the 2 datasets into separate dataframes
private_df = pd.read_csv('Private_Schools.csv')
public_df = pd.read_csv('Public_Schools.csv')
```

```python
#display uncleaned private_df
private_df
```

```python
#display uncleaned public_df
public_df
```

```python
#prepare for table concatenation by introducing a field called is_private
public_df["is_private"] = 0
private_df["is_private"] = 1
```

```python
#concatenate the two tables for easier querying
df_schools = pd.concat([public_df, private_df])
```

```python
#rename columns according to ER diagram
df_schools = df_schools.rename(columns={'School_id':'Objectid', 'Start_grade' : 'st_grade'})
df_schools = df_schools.rename(columns={'OBJECTID':'Objectid', 'ST_GRADE' : 'st_grade', 'NAME':'name', 'ADDRESS':'address',
                                        'CITY':'city', 'STATE':'state', 'TYPE':'type', 'POPULATION':'population',
                                        'END_GRADE':'end_grade'})
```

```python
#only keep the columns we need
df_schools_cleaned = df_schools[['Objectid', 'name', 'address','city', 'state', 'type', 'population', 'st_grade', 'end_grade', 'is_private']]
```

```python
#display uncleaned concatenated table of public and private schools
df_schools_cleaned
```

```python
#break down the table into 2, school_location and school_stats, to achieve BCNF
school_location_df = df_schools_cleaned[['Objectid', 'name', 'address', 'city', 'state', 'is_private']]
```

```python
#display cleaned schoolLocation (to be outputed)
school_location_df
```

```python
#break down the table into 2, school_location and school_stats, to achieve BCNF
school_stats_df = df_schools_cleaned[['Objectid', 'type', 'population', 'st_grade', 'end_grade', 'is_private']]
```

```python
#display cleaned schoolStats (to be outputed)
school_stats_df
```

```python
school_location_df.to_csv('SchoolLocation.csv')
```

```python
  
school_stats_df.to_csv('SchoolStats.csv')
```

# **MICHELIN STAR RESTAURANTS**

```python
#store the csv file in a dataframe
df_mich = pd.read_csv('michelin_star.csv')
```

```python
#display uncleaned df_mich
df_mich
```

```python
#drop unneeded columns
df_mich = df_mich.drop(columns = ['blurb','image', 'chef', 'url', 'url2'])

#convert string representation of stars to ints
df_mich.loc[df_mich["michelin_award"] == "ONE_STAR", "michelin_award"] = 1
df_mich.loc[df_mich["michelin_award"] == "TWO_STARS", "michelin_award"] = 2
df_mich.loc[df_mich["michelin_award"] == "THREE_STARS", "michelin_award"] = 3

# Merge cuisine1 and cuisine2 
df_mich = df_mich.fillna(" ")
df_mich['cuisine'] = df_mich['cuisine1'] + ', ' + df_mich['cuisine2']

#drop after merging
df_mich = df_mich.drop(columns = ['cuisine1', 'cuisine2'])

#rename according to ER diagram
df_mich = df_mich.rename(columns={'michelin_award' : 'num_michelin_stars', 'lat' : 'latitude', 'lon':'longitude'})

# Put defaults in for non primary keys columns
df_mich['num_michelin_stars'] = df_mich['num_michelin_stars'].fillna(1)


df_mich['country'] = df_mich['country'].fillna("United States")
df_mich['cuisine'] = df_mich['cuisine'].fillna("American")

# Drop empty rows now as it will eliminate nans in primary key columns
df_mich = df_mich.dropna()
```

```python
#display cleaned df_mich
df_mich
```

```python
# Save dataset in csv files
df_mich.to_csv('MichelinRestaurants.csv')
```

# **AIRBNB 1**

```python
#store the csv file in a dataframe
airbnb_df = pd.read_csv('airbnb1_uncleaned.csv')
```

```python
#display uncleaned airbnb_df
airbnb_df
```

```python
#drop unnecessary columns
airbnb_df = airbnb_df.drop(['latitude', 'host_id', 'longitude', 'calculated_host_listings_count', 'availability_365', 'last_review', 'reviews_per_month', 'name'], axis=1)
airbnb_df = airbnb_df.drop(['host_name', 'minimum_nights', "number_of_reviews"], axis=1)

#drop null rows
airbnb_df= airbnb_df.dropna() 

#remove duplicates
airbnb_df.drop_duplicates(inplace=True)

#rename according to the ER diagram
airbnb_df = airbnb_df.rename(columns={'neighbourhood_group':'neighborhood_group', 'neighborhood' : 'neighborhood', 'price' : 'price_per_night'})

#set column order to match with DDL statements
col_order = ['id', 'city', 'neighborhood_group', 'neighborhood', 'room_type', 'price_per_night']

#filter out only the columns in col_order
airbnb_df = airbnb_df.filter(col_order)
```

```python
#strip city column and keep string at index 0
airbnb_df['city'] = airbnb_df['city'].str.split(',').str[0]
```

```python
#display cleaned airbnb_df
airbnb_df
```

```python
airbnb_df.to_csv('AIRBNB.csv')
```

# **AIRBNB 2**

```python
#read csv into dataframe and use error_bad_lines to deal with the jaggered
#nature of the csv
airbnb2_df = pd.read_csv("airbnb2_uncleaned.csv", delimiter=';', error_bad_lines=False)
```

```python
#display uncleaned airbnb2_df
airbnb2_df
```

```python
#only keep the columns we need
airbnb2_df_chopped = airbnb2_df[['ID', 'City', 'Neighbourhood', 'Room Type', 'Price']]

#rename according to the ER diagram
airbnb2_df_chopped.rename(columns={'ID': 'id', 'Neighbourhood':'neighborhood_group', 'Room Type': 'room_type',
                                'Price': 'price_per_night', 'City': 'city'}, inplace=True)

#drop nulls in the columns we are keeping
airbnb2_df_chopped.dropna(subset=['id', 'neighborhood_group', 'room_type', 'price_per_night', 'city'], inplace=True)
```

```python
#strip city column and keep string at index 0
airbnb2_df_chopped['city'] = airbnb2_df_chopped['city'].str.split('_').str[0]
```

```python
#display cleaned airbnb2_df
airbnb2_df_chopped
```

```python
airbnb2_df_chopped.to_csv('AIRBNB2.csv')
```
